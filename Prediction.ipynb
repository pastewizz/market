```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# read the dataset you have first
data = pd.read_csv('marketdata.csv')

# Devide the data into input and output
X = data[['AdSpend', 'CompetitorActivity']]
y = data['MarketDemand']

# create the model
model = LinearRegression()

#train the model with the dataset
model.fit(X, y)

# Predict market demand
data['PredictedDemand'] = model.predict(X)

# Print predictions
print(data[['Month', 'MarketDemand', 'PredictedDemand']])

# Visualize actual vs predicted
plt.plot(data['Month'], data['MarketDemand'], label='Actual')
plt.plot(data['Month'], data['PredictedDemand'], label='Predicted', linestyle='--')
plt.xlabel('Month')
plt.ylabel('Market Demand')
plt.title('Market Demand Prediction')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
